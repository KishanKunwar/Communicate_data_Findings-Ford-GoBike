## Dataset Overview
The dataset for this project is called "Ford GoBike Data". This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. It include 23911 entries and 14 columns such as duration_sec,	start_time,	end_time, start_station_name, start_station_latitude, start_station_longitude,  end_station_name, end_station_latitude, end_station_longitude, b user_type 	etc before cleaning the dataset. 

### gathering dataset


```python
#Importing Necessary Package for the Project
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb 
%matplotlib inline
```

```python
# Reading the dataset
df = pd.read_csv("fordgobike-tripdata.csv")
df.head(2)
```

```python
#Getting Information about dataset
df.info()
```

```python
#Making a copy of original dataset
df_copy = df.copy()
```

```python
#Checking for null values
df_copy.isnull().sum()
```

```python
#Removing Null Values 
df_copy.dropna(inplace = True)
```

```python
#Rechecking For null values if present after removing 
df_copy.isnull().sum()
```

```python
#Checking start_station_name values
df_copy["start_station_name"].value_counts()
```

```python
# Converting it into category datatype
df_copy["start_station_name"]=df_copy["start_station_name"].astype('category')
```

```python
#Checking values 
df_copy["end_station_name"].value_counts()
```

```python
# Converting it into category datatype
df_copy["end_station_name"]=df_copy["end_station_name"].astype('category')
```

```python
# Checking values 
df_copy["user_type"].value_counts()
```

```python
# Converting it into category datatype
df_copy["user_type"] = df_copy["user_type"].astype('category')
```

```python
# geting information
df_copy.info()
```

```python
# Converting to datetime datatype
df_copy["start_time"]= pd.to_datetime(df_copy["start_time"])
df_copy["end_time"]= pd.to_datetime(df_copy["end_time"])
```

```python
# Extracting year, month and day from start_time
df_copy["starting_year"]= df_copy["start_time"].dt.year
df_copy["starting_month"]= df_copy["start_time"].dt.month
df_copy["starting_day"]= df_copy["start_time"].dt.day
```

```python
# Extracting year, month and day from end_time
df_copy["ending_year"]= df_copy["end_time"].dt.year
df_copy["ending_month"]= df_copy["end_time"].dt.month
df_copy["ending_day"]= df_copy["end_time"].dt.day

```

```python
# Converting Datatype
df_copy["starting_year"]=df_copy["starting_year"].astype('category')
df_copy["starting_month"]=df_copy["starting_month"].astype('category')
df_copy["starting_day"]=df_copy["starting_day"].astype('int')

df_copy["ending_year"]=df_copy["ending_year"].astype('category')
df_copy["ending_month"]=df_copy["ending_month"].astype('category')
df_copy["ending_day"]=df_copy["ending_day"].astype('int')
```

```python
# Checking values 
df_copy["bike_share_for_all_trip"].value_counts()
```

```python
# Converting it into category datatype
df_copy["bike_share_for_all_trip"]=df_copy["bike_share_for_all_trip"].astype('category')
```

```python
df_copy.head(2)
```

```python
# getting latitidue and logitude difference 
df_copy["latitude_diff"] = df_copy["end_station_latitude"]- df_copy["start_station_latitude"]  
df_copy["longitude_diff"] = df_copy["end_station_longitude"]- df_copy["start_station_longitude"]
```

```python
# Creating a categorical column base on above difference
df_copy["north_south"]= df_copy.latitude_diff.apply(lambda x : "North" if x > 0 else "South")
df_copy["east_west"]= df_copy.longitude_diff.apply(lambda x : "East" if x > 0 else "West")
```

```python
# Converting it into category datatype
df_copy["north_south"] = df_copy["north_south"].astype('category')
df_copy["east_west"] = df_copy["east_west"].astype('category')
```

```python
df_copy.head(2)
```

```python
df_copy.info()
```

- Its seems like day of the month doest affect the duration of ride for bith tyoe of user as all seems similar for each day

```python
#Plotting a scatter plot using shape to add a categorical variable
plt.figure(figsize = [17, 10])

ttype_markers =[["Subscriber","o"],
               ["Customer","X"]]

for ttype,marker in ttype_markers:
    plot_data = df_copy.loc[df_copy["user_type"]==ttype]
    sb.regplot(data = plot_data , x="starting_day", y="duration_sec",x_jitter=0.37, scatter_kws={'alpha':.3} , fit_reg= False,  marker=marker)

plt.legend(["Subscriber","Customer"])
```

## To conclude the analysis , duration for riding bike is irrespective to the days of the month, type of the user . However most people tends to go south-west part of the area of San Francisco Bay and average 802 sec is spent by an average rider.
